## batch size 임베딩 모델 선택할때 

```python
from src.tokencounter import count_tokens
from src.loader import load_pdf_paths
from src.loader import extract_text_PyMuPDF
from src.cleaner import clean_text

# 전체 토큰 수 저장용
token_lengths_all = []

pdf_dir = "ingestion/data/vector_store/pdf"
pdf_paths = load_pdf_paths(pdf_dir)

# 평균 토큰 계산만을 위한 루프
for pdf in pdf_paths:
    raw_text = extract_text_PyMuPDF(pdf)
    cleaned_text = clean_text(raw_text)
    
    token_len = count_tokens(cleaned_text)
    token_lengths_all.append(token_len)

# 평균 토큰 수 계산
avg_token_len_all = sum(token_lengths_all) // len(token_lengths_all)

# 안전한 최대 토큰 한도 (OpenAI 기준)
safe_token_limit = 8000

# 추천 배치 사이즈 계산
suggested_batch_size = max(1, safe_token_limit // (avg_token_len_all + 1))

print(f"📊 평균 토큰 수: {avg_token_len_all}")
print(f"📦 추천 batch size: {suggested_batch_size}")
```
